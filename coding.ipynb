```python
import matplotlib.pyplot as plt
import numpy as np

cities = ['Montreal', 'Ottawa']
cities2 = ['Quebec(Montreal)', 'Ontario(Ottawa)']
city_population = [1.76, 1.01]
gma_population = [4.29, 1.49]
two_hour_drive_population = [5, 5] 
one_day_drive_population = [8, 8]
regional_gdp = [400.9, 800.4]
population_growth = [0.72, 0.98]
economic_growth = [1.7, 1.7]
```

```python
x = np.arange(len(cities))  
width = 0.2  

fig, ax1 = plt.subplots(figsize=(10, 6))

ax2 = ax1.twinx()

rects1 = ax1.bar(x - width, city_population, width, label='City Population(Million)', color=colorlist[7])
rects2 = ax1.bar(x, gma_population, width, label='GMA Population Million)', color=colorlist[6])
rects3 = ax2.bar(x + width, regional_gdp, width, label='Regional/Province GDP(Billion)', color=colorlist[5])
#add number on bar
def add_labels_to_bars(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Call the function for each set of bars
add_labels_to_bars(rects1, ax1)
add_labels_to_bars(rects2, ax1)
add_labels_to_bars(rects3, ax2)  # Assuming ax2 is used for the GDP bars


ax1.set_ylabel('Population(Million)')
ax2.set_ylabel('GDP(Billion)')
ax1.set_title('Population & GDP Comparison: Montreal vs. Ottawa (CAD)')
ax1.set_xticks(x)
ax1.set_xticklabels(cities)
ax1.set_ylim([0, 6])
ax2.set_ylim([0,1100])

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 6))

rects1 = ax.bar(x, population_growth, width, label='Population Growth (%)', color= colorlist[8])
# Add text labels on the top of the bars
for rect in rects1:
    height = rect.get_height()
    ax.annotate(f'{height:.2f}%',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
ax.set_ylabel('Growth Rate (%)')
ax.set_title('Population Growth Rate: Montreal vs Ottawa')
ax.set_xticks(x)
ax.set_xticklabels(cities)
ax.legend()
ax.set_ylim([0,1.2])

fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 6))

rects2 = ax.bar(x, economic_growth, width, label='Economic Growth (%)', color= colorlist[9])
# Add text labels on the top of the bars
for rect in rects2:
    height = rect.get_height()
    ax.annotate(f'{height:.2f}%',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
ax.set_ylabel('Growth Rate (%)')
ax.set_title('Economic Growth Rate(Regional): Quebec (Montreal) vs. Ontario (Ottawa)')
ax.set_xticks(x)
ax.set_xticklabels(cities2)
ax.legend()
ax.set_ylim([0,2])
fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
years = ['2017', '2018', '2019', '2020']
montreal_gdp = [211625, 221893, 233996, 228707]
ottawa_gdp = [67241, 70086, 73878, 74821]

# Create figure and plot space
plt.figure(figsize=(10, 6))

# Plotting Montreal GDP
plt.plot(years, montreal_gdp, color=colorlist[3], marker='o', label='Montreal')
# Adding data labels for Montreal
for i, txt in enumerate(montreal_gdp):
    plt.annotate(txt, (years[i], montreal_gdp[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Plotting Ottawa GDP
plt.plot(years, ottawa_gdp, color=colorlist[2], marker='o', label='Ottawa')
# Adding data labels for Ottawa
for i, txt in enumerate(ottawa_gdp):
    plt.annotate(txt, (years[i], ottawa_gdp[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Title and labels
#plt.title('GDP Growth from 2017 to 2020')
plt.xlabel('Year')
plt.ylabel('GDP in Million $')

plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import numpy as np

# Function to calculate growth rate
def calculate_growth_rate(data):
    return np.diff(data) / data[:-1] * 100

# Calculating growth rates
montreal_growth = calculate_growth_rate(montreal_gdp)
ottawa_growth = calculate_growth_rate(ottawa_gdp)

# Adjusting years for growth rate calculation (since the first year has no previous year to compare with)
growth_years = years[1:]  # From 2018 to 2020

# Create figure and plot space
plt.figure(figsize=(10, 6))

# Plotting Montreal growth rate
plt.plot(growth_years, montreal_growth, color=colorlist[3], marker='o', label='Montreal Growth Rate')
# Adding data labels for Montreal growth rate
for i, txt in enumerate(montreal_growth):
    plt.annotate(f"{txt:.2f}%", (growth_years[i], montreal_growth[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Plotting Ottawa growth rate
plt.plot(growth_years, ottawa_growth, color=colorlist[2], marker='o', label='Ottawa Growth Rate')
# Adding data labels for Ottawa growth rate
for i, txt in enumerate(ottawa_growth):
    plt.annotate(f"{txt:.2f}%", (growth_years[i], ottawa_growth[i]), textcoords="offset points", xytext=(0,-20), ha='center')

# Title and labels
#plt.title('Annual GDP Growth Rate from 2017 to 2020')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
